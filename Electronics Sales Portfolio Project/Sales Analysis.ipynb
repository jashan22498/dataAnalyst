# Which city had the maximum sales?

```python
# Adding a 'City' column
```

```python
all_months_data.head()
```

```python
# 1. by using apply() method
all_months_data['City'] = all_months_data['Purchase Address'].apply(lambda x: x.split(',')[1][1:])
```

```python
# 2.
def get_city(address):
    return address.split(',')[1][1:]

all_months_data['City'] = all_months_data['Purchase Address'].apply(lambda x: get_city(x))
```

```python
all_months_data
```

```python
all_months_data.loc[0, 'City']
```

```python
# 'Portland' is in two states ('OR' and 'ME')
```

```python
# Adding state to 'City' column

def get_state(address):
    return address.split(',')[2][1:3]

all_months_data.City = all_months_data.City + ' (' + all_months_data['Purchase Address'].apply(lambda x: get_state(x)) + ')'
```

# Maximum number of orders are placed between 7pm and 8pm or 12pm and 1pm; these two are peak hours for ads to be placed

```python
# line graph to show orders count per hour of the day 
```

```python
figure(figsize = (10, 8), dpi = 80)

plt.title('Line Graph - Hourly Orders')

plt.plot(np.sort(hours), hourlyOrders['Order ID'])

plt.xticks(hours)
plt.grid()

plt.xlabel('Hour')
plt.ylabel('Number of Orders')

plt.show()
```

```python
pd.DataFrame({'Hour': np.sort(hours), 'Orders Count': hourlyOrders['Order ID']}).to_csv('./hourlyOrders.csv', index = False)
```
